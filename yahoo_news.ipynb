```python
import urllib2
resp = urllib2.urlopen("https://tw.news.yahoo.com/")
from IPython.display import HTML
resp = resp.read().decode('utf8')
from pyquery import PyQuery
S = PyQuery(resp)
print len(S('a')) # Amount of "a" tag
```
